```python
#Filtro por columnas
filtro3 = dataframe.iloc[:, 4:7] #Igual que el anterior
filtro3 #Muestra las columnas de la 4 a 7, en este caso como el archivo solo tiene hasta columna 6 se muestran solo 3
```

```python
dataframe = pd.read_csv("file.csv",index_col=2) #Quita columna 2 del dataframe "entidad federativa"
dataframe #Con este nuevo dataframe se trabaja para los siguientes filtros
```

**Filtros Basicos PARA VAIRABLES CUANTITATIVAS**

```python
#Filtro por comparación
filtro6 = dataframe[dataframe["Periodo"]>2006]
filtro6 #Muestra solo los datos pertientes a Periodos mayores a 2006
```

```python
#Convertir archivo filtrado a csv
filtro6.to_csv("Entregable6.csv")

#descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable6.csv")
```

**Filtros por Argumentos Logicos**

```python
#Filtro Y (Filtro And)
filtro7 = dataframe[(dataframe["Periodo"] > 2006) & (dataframe["Sexo"] == "Mujeres")]
filtro7 #Muestra solo los datos pertinentes a mujeres que son de periodo mayor a 2006
```

```python
#Convertir archivo filtrado a csv
filtro7.to_csv("Entregable7.csv")

#descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable7.csv")
```

```python
#Filtro o (Filtro or)
filtro8 = dataframe[(dataframe["Periodo"] > 2006) | (dataframe["Sexo"] == "Mujeres")]
filtro8 #Muestra solo todos los datos pertinentes a mujeres y todos los datos con periodo mayor a 2006
```

```python
#Convertir archivo filtrado a csv
filtro8.to_csv("Entregable8.csv")

#descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable8.csv")
```

```python
#Filtro no (Filtro not)

filtro9 = dataframe[~(dataframe["Periodo"] > 2006) & ~(dataframe["Sexo"] == "Mujeres")]
filtro9 #Muestra todos los datos que no son pertinentes a periodo mayor que 2006 y que no son pertinentes a mujeres
```

```python
#Convertir archivo filtrado a csv
filtro9.to_csv("Entregable9.csv")

#descargar archivo filtrado en csv
from google.colab import files

files.download("Entregable9.csv")
```
